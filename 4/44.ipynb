```python
# Perceptron Learning Law with Decision Region (Simple Version)

import numpy as np
import matplotlib.pyplot as plt

# Input data (2 features)
X = np.array([
    [1, 1],
    [2, 1],
    [4, 3],
    [5, 4]
])

# Target output
# Class -1 = Blue
# Class 1 = Red
y = np.array([-1, -1, 1, 1])

# Initial weights and bias
w = np.zeros(2)
b = 0
lr = 1

# Training using Perceptron Learning Law
for epoch in range(10):
    for i in range(len(X)):

        net = np.dot(X[i], w) + b

        if net >= 0:
            output = 1
        else:
            output = -1

        # Update weights if wrong prediction
        if output != y[i]:
            w = w + lr * y[i] * X[i]
            b = b + lr * y[i]

# Create decision region
x1_min = 0
x1_max = 6
x2_min = 0
x2_max = 5

xx, yy = np.meshgrid(np.arange(x1_min, x1_max, 0.1),
                     np.arange(x2_min, x2_max, 0.1))

# Predict all grid points
Z = []

for i in range(len(xx.ravel())):
    point = np.array([xx.ravel()[i], yy.ravel()[i]])
    net = np.dot(point, w) + b

    if net >= 0:
        Z.append(1)
    else:
        Z.append(-1)

Z = np.array(Z).reshape(xx.shape)

# Plot graph
plt.contourf(xx, yy, Z, alpha=0.4, cmap="bwr")
plt.scatter(X[:,0], X[:,1], c=y, cmap="bwr", s=100, edgecolors="black")

plt.title("Perceptron Decision Region")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```
