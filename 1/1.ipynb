```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,100)
x
```

```python
# step 
def step(x):
    return np.where(x>=0,1,0)
```

```python
#sigmoid
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
# tahn
def tanh(x):
    return np.tanh(x)
```

```python
#relu 
def relu(x):
    return np.maximum(0,x)
```

```python
# leaky relu 
def lerelu(x):
    return np.where(x>0,x,0.01*x)
```

```python
# softmax
def softmax(x):
    e=np.exp(x-np.max(x))
    return e/e.sum()
```

```python
# create graph window
plt.figure(figsize=(15,10))
```

```python
plt.subplot(3,3,1)
plt.plot(x,step(x))
plt.title("step function")
```

```python
plt.subplot(3,3,2)
plt.plot(x ,sigmoid(x))
plt.title("Sigmoid function")
```

```python
plt.subplot(3,3,3)
plt.plot(x ,tanh(x))
plt.title("tanh function")
```

```python
plt.subplot(3,3,4)
plt.plot(x ,relu(x))
plt.title("relu function")
```

```python
plt.subplot(3,3,5)
plt.plot(x ,lerelu(x))
plt.title("leaky relu function")
```

```python
# softmax
y= np.array([2,1,0.1])
s= softmax(y)
plt.subplot(3,3,7)
plt.bar(range(len(s)), y)
plt.title("Softmax")
```
