```python

# Plot Activation Functions used in Neural Networks

import numpy as np
import matplotlib.pyplot as plt

# Input values from -10 to 10
x = np.linspace(-10, 10, 100)

# Activation Functions
def step(x):
    return np.where(x >= 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x):
    return np.where(x > 0, x, 0.01 * x)

def elu(x):
    return np.where(x > 0, x, np.exp(x) - 1)

def softmax(x):
    e = np.exp(x - np.max(x))
    return e / e.sum()

# Create Graph Window
plt.figure(figsize=(15,10))

# 1 Step Function
plt.subplot(3,3,1)
plt.plot(x, step(x))
plt.title("Step Function")

# 2 Sigmoid
plt.subplot(3,3,2)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid")

# 3 Tanh
plt.subplot(3,3,3)
plt.plot(x, tanh(x))
plt.title("Tanh")

# 4 ReLU
plt.subplot(3,3,4)
plt.plot(x, relu(x))
plt.title("ReLU")

# 5 Leaky ReLU
plt.subplot(3,3,5)
plt.plot(x, leaky_relu(x))
plt.title("Leaky ReLU")

# 6 ELU
plt.subplot(3,3,6)
plt.plot(x, elu(x))
plt.title("ELU")

# 7 Softmax Example
sample = np.array([2,1,0.1])
y = softmax(sample)

plt.subplot(3,3,7)
plt.bar(range(len(sample)), y)
plt.title("Softmax")

plt.tight_layout()
plt.show()
```
