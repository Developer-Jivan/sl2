```python
# andnot gate mcpitt

# step function
def step(net, threshold):
    if net>=threshold:
        return 1
    else :
        return 0

inputs=[(0,0),(0,1),(1,0),(1,1)]

w1=1
w2=-1
th=1

for x1 , x2 in inputs:
    temp=(x1*w1)+(x2*w2)
    output=step(temp,th)
    print("x1: ",x1,"x2: ",x2,"output: ",output)

```
