```python
# McCulloch-Pitts Neural Network for ANDNOT Function
# Output = X1 AND (NOT X2)

# Step Activation Function
def step_function(net, threshold):
    if net >= threshold:
        return 1
    else:
        return 0

# All possible input combinations
inputs = [(0,0), (0,1), (1,0), (1,1)]

# Weights and Threshold
w1 = 1      # for X1
w2 = -1     # for X2 (NOT effect)
threshold = 1

print("X1  X2  Output")

# Process each input
for x1, x2 in inputs:
    net_input = (x1 * w1) + (x2 * w2)
    output = step_function(net_input, threshold)
    print(x1, " ", x2, "   ", output)
```
