```python
# Artificial Neural Network using Forward Propagation and Back Propagation
# Problem: XOR Logic Gate

import numpy as np

# -----------------------------------------
# Step 1: Input data
# -----------------------------------------
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

# Expected Output
Y = np.array([[0],
              [1],
              [1],
              [0]])

# -----------------------------------------
# Step 2: Initialize weights and bias
# -----------------------------------------
np.random.seed(1)   # Same random values every run

# Input layer -> Hidden layer weights
W1 = np.random.rand(2, 2)

# Hidden layer -> Output layer weights
W2 = np.random.rand(2, 1)

# Bias values
b1 = np.random.rand(1, 2)
b2 = np.random.rand(1, 1)

# -----------------------------------------
# Step 3: Activation Function
# -----------------------------------------
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of Sigmoid Function
def sigmoid_derivative(x):
    return x * (1 - x)

# -----------------------------------------
# Step 4: Training
# -----------------------------------------
for i in range(10000):

    # -------- Forward Propagation --------

    # Hidden layer output
    H = sigmoid(np.dot(X, W1) + b1)

    # Final output
    O = sigmoid(np.dot(H, W2) + b2)

    # -------- Error --------
    error = Y - O

    # -------- Back Propagation --------

    # Output layer gradient
    dO = error * sigmoid_derivative(O)

    # Hidden layer gradient
    dH = np.dot(dO, W2.T) * sigmoid_derivative(H)

    # -------- Update Weights --------

    W2 = W2 + np.dot(H.T, dO) * 0.1
    W1 = W1 + np.dot(X.T, dH) * 0.1

    # -------- Update Bias --------

    b2 = b2 + dO.sum(axis=0) * 0.1
    b1 = b1 + dH.sum(axis=0) * 0.1

# -----------------------------------------
# Step 5: Output
# -----------------------------------------
print("Predicted Output:")
print(O.round())
```
